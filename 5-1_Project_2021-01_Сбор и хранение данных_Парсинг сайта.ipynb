```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
req = requests.get('https://code.s3.yandex.net/learning-materials/data-analyst/festival_news/index.html')
#print(req.text) # вывод содержимого веб-страницы
soup = BeautifulSoup(req.text, 'lxml')
#print(soup)
table = soup.find('table', attrs={'id': 'best_festivals'})
#print(table.text) # контент без тега

columns=[]
for row in table.find_all('th'): 
    columns.append(row.text)
#print(columns) 

data = []
for row in table.find_all('tr'): 
    if not row.find_all('th'): 
        data.append([element.text for element in row.find_all('td')])
#print(data) 

festivals = pd.DataFrame(data=data, columns=columns)
festivals
```
