### 2.1.1 Рассчёт DAU, WAU и MAU <a id="shag2.1.1"></a>  
[К содержанию](#soder)  

•	2.1.1. Рассчитайте DAU, WAU и MAU. Вычислите средние значения этих метрик за весь период, а также постройте график изменения метрик во времени;

```python
visits['year'] = visits['start_ts'].dt.year
visits['month'] = visits['start_ts'].dt.month
visits['week'] = visits['start_ts'].dt.week
visits['date'] = visits['start_ts'].dt.date
visits
```

```python
mau = visits.groupby(['year', 'month']).agg({'uid':'nunique'}).reset_index()
wau = visits.groupby(['year', 'week']).agg({'uid':'nunique'}).reset_index()
dau = visits.groupby(['date']).agg({'uid':'nunique'}).reset_index()
```

```python
mau_mean = mau['uid'].mean()
wau_mean = wau['uid'].mean()
dau_mean = dau['uid'].mean()

print('Среднее значение MAU =', int(round(mau_mean)) )
print('Среднее значение WAU =', int(round(wau_mean)) )
print('Среднее значение DAU =', int(round(dau_mean)) )
```

### 2.1.4 Рассчёт Retention Rate <a id="shag2.1.4"></a>  
[К содержанию](#soder)  

•	2.1.4. Рассчитайте Retention Rate, применяя когортный анализ.

**По-скольку весь период времени в нашей таблице 1 год, когорты возьмём кратные 1 месяцу.**

```python
visits.head(3)
```

```python
first_visit_month_uid = visits.groupby('uid')['start_ts'].min().astype('datetime64[M]')
first_visit_month_uid.name = 'first_visit_month'
print(first_visit_month_uid.head()) 
```

```python
visits = visits.join(first_visit_month_uid, on='uid')
visits.head()
```

```python
visits['visit_month'] = visits['start_ts'].astype('datetime64[M]')
visits.sample(6)
```

```python
visits['cohort_lifetime'] = visits['visit_month'] - visits['first_visit_month']
visits['cohort_lifetime'] = visits['cohort_lifetime'] / np.timedelta64(1,'M')
visits['cohort_lifetime'] = visits['cohort_lifetime'].round().astype('int') 
```

```python
# visits.sample(15)
visits['cohort_lifetime'].unique()
```

<div style="border:solid green 2px; padding: 20px"> 
    
**Комментарий ревьюера**

**Сгруппируем данные по когорте и lifetime. Посчитаем для каждой когорты количество активных пользователей на определённый «месяц жизни»:**

```python
cohorts = visits.groupby(['first_visit_month','cohort_lifetime']).agg({'uid':'nunique'}).reset_index()
cohorts.sample(6)
```

**Чтобы найти Retention Rate, нужно сперва получить число пользователей, изначально бывших в когорте, и на него разделить число пользователей в каждый следующий месяц.  
Найдём исходное количество пользователей в когорте. Возьмём их число на нулевой месяц:**

```python
initial_visiters_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_visit_month','uid']]
initial_visiters_count = initial_visiters_count.rename(columns={'uid':'visiters'})
initial_visiters_count
```

**Объединим `cohorts` и `initial_visiters_count` и наконец, рассчитаем Retention Rate. Разделим количество активных пользователей в каждый месяц на исходное число пользователей в когорте:**

```python
cohorts = cohorts.merge(initial_visiters_count,on='first_visit_month')
cohorts['retention'] = cohorts['uid']/cohorts['visiters'] 
cohorts.sample(8)
```

**Представим все ввиде сводной таблице и нарисуем тепловую карту**

```python
cohorts['first_visit_month'] = cohorts['first_visit_month'].dt.strftime('%Y-%m')
```

```python
retention_pivot = cohorts.pivot_table(index='first_visit_month',columns='cohort_lifetime',values='retention',aggfunc='sum') 
del retention_pivot[0] # Удаляем так как здесь 100% и это мешает цветовой гамме
retention_pivot
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
sns.set(style='white')
plt.figure(figsize=(15, 9))
plt.title('Cohorts: Visiter Retention')
sns.heatmap(retention_pivot, annot=True, fmt='.2%', linewidths=1, linecolor='g')
plt.xlabel('МЕСЯЦ "ЖИЗНИ"')
plt.ylabel('КОГОРТЫ'); 
```

### 3.1.3 Retention Rate (по устройствам) <a id="shag3.1.3"></a>  
[К содержанию](#soder) 

```python
cohorts_touch = visits[visits['category_uid_device']=='touch']. \
                groupby(['first_visit_month','cohort_lifetime']).agg({'uid':'nunique'}).reset_index()
display("cohorts_touch",cohorts_touch.sample(3))
print()
cohorts_desktop = visits[visits['category_uid_device']=='desktop']. \
                groupby(['first_visit_month','cohort_lifetime']).agg({'uid':'nunique'}).reset_index()
display('cohorts_desktop',cohorts_desktop.sample(3))
print()
cohorts_td = visits[visits['category_uid_device']=='td']. \
                groupby(['first_visit_month','cohort_lifetime']).agg({'uid':'nunique'}).reset_index()
display("cohorts_td",cohorts_td.sample(3))
```

**Чтобы найти Retention Rate, нужно сперва получить число пользователей, изначально бывших в когорте, и на него разделить число пользователей в каждый следующий месяц.  
Найдём исходное количество пользователей в когорте. Возьмём их число на нулевой месяц:**

```python
initial_n_visiters_touch = cohorts_touch[cohorts_touch['cohort_lifetime'] == 0][['first_visit_month','uid']]
initial_n_visiters_touch = initial_n_visiters_touch.rename(columns={'uid':'visiters'})
display('initial_n_visiters_touch',initial_n_visiters_touch.sample(3))
print()
initial_n_visiters_desktop = cohorts_desktop[cohorts_desktop['cohort_lifetime'] == 0][['first_visit_month','uid']]
initial_n_visiters_desktop = initial_n_visiters_desktop.rename(columns={'uid':'visiters'})
display('initial_n_visiters_desktop',initial_n_visiters_desktop.sample(3))
print()
initial_n_visiters_td = cohorts_td[cohorts_td['cohort_lifetime'] == 0][['first_visit_month','uid']]
initial_n_visiters_td = initial_n_visiters_td.rename(columns={'uid':'visiters'})
display('initial_n_visiters_td',initial_n_visiters_td.sample(3))
```

**Объединим `cohorts` и `initial_visiters_count` и наконец, рассчитаем Retention Rate. Разделим количество активных пользователей в каждый месяц на исходное число пользователей в когорте:**

```python
cohorts_touch = cohorts_touch.merge(initial_n_visiters_touch, on='first_visit_month')
cohorts_touch['retention'] = cohorts_touch['uid']/cohorts_touch['visiters'] 
display("cohorts_touch", cohorts_touch.sample(3))
print()
cohorts_desktop = cohorts_desktop.merge(initial_n_visiters_desktop, on='first_visit_month')
cohorts_desktop['retention'] = cohorts_desktop['uid']/cohorts_desktop['visiters'] 
display("cohorts_desktop", cohorts_desktop.sample(3))
print()
cohorts_td = cohorts_td.merge(initial_n_visiters_td, on='first_visit_month')
cohorts_td['retention'] = cohorts_td['uid']/cohorts_td['visiters'] 
display("cohorts_td", cohorts_td.sample(3))
```
